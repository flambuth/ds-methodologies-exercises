```python
def plot_variable_pairs(dataframe):
    train, test = split_scale.split_my_data(dataframe)
    g = sns.pairplot(train, hue='tenure_years')
    g.map_diag(plt.hist)
    g.map_offdiag(plt.scatter);
```
