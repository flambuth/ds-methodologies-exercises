```python
df = pd.read_csv("billboard.csv", encoding="mac_latin2")
```

```python
df.head()
```

```python
id_vars = ['year', 'artist.inverted', 'track', 'time', 'genre', 'date.entered', 'date.peaked']
```

```python
df_melted = pd.melt(frame=df,id_vars=id_vars, var_name="week", value_name="rank")
```

```python
df_melted.head()
```

```python
df_melted['week'] = df_melted['week'].str.extract('(\d+)', expand=False).astype(int)
```

```python
df_melted.info()
```

```python
df['date'] = pd.to_datetime(df['date.entered']) + pd.to_timedelta(df['week'], unit='w') - pd.DateOffset(weeks=1)
```
